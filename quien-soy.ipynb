<h1>Importamos librerías</h1>

```python
### LIBRERIAS A UTILIZAR ###
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>Leemos el excel</h1>

```python
quien = pd.read_excel("Quien.xlsx")
quien.head()
```

```python
### Eliminamos la primera columna
quien = quien.drop('No.', axis=1)
quien.head()
```

<h1>Analizar los datos</h1>

```python
quien_num = quien.replace(['SI','NO'], [1,0])
quien_num = quien_num.replace(['SÍ','No'], [1,0])
quien_num = quien_num.replace('Si',1)
```

<h1>Importamos los algoritmos de machine learning</h1>

```python
### ALGORITMOS DE MACHINE LEARNING ###
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.neighbors import KNeighborsClassifier
```

<h1>Obtener datos de entrenamiento y prueba</h1>

```python
#Calculamos sin los datos faltantes para que no afecte al modelo
X = np.array(quien_num.iloc[0:23].drop(['Personaje','Sexo'],axis=1))
Y = np.array(quien_num.iloc[0:23]['Sexo'])

#X_train --> Y_train   entrenar mi red  80%
#X_test --> Y_test     validar --> predecir  20%

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
print("Son {} datos para entrenamiento y {} datos de prueba".format(X_train.shape[0], X_test.shape[0]))
print("Son {} etiquetas para entrenamiento y {} etiquetas de prueba".format(Y_train.shape[0], Y_test.shape[0]))
```

<h1>Aplicamos Regresión Logistica</h1>

```python
modelo =  LogisticRegression()
modelo.fit(X_train,Y_train)
```

<h3>Predecir valores faltantes</h3>

```python
prueba = quien_num.iloc[[24,25]]
prueba = prueba.drop(['Personaje','Sexo'],axis=1)
prueba= prueba.to_numpy()
print(prueba)

faltantes = modelo.predict(prueba)
print(faltantes)
print(quien_num.iloc[[24,25]]['Personaje'])
```

<h3>Obtener Score de predicción</h3>

```python
print("Presición Regresión Logistica: {}".format(modelo.score(X_train, Y_train)))
```

<h1>Aplicamos SVM o Máquinas de vectores de soporte</h1>

```python
modelo = SVC()
modelo.fit(X_train, Y_train)
```

<h3>Predecir valores faltantes</h3>

```python
faltantes = modelo.predict(prueba)
print(faltantes)
print(quien_num.iloc[[24,25]]['Personaje'])
```

<h3>Obtener Score de predicción</h3>

```python
print("Presición Regresión Logistica: {}".format(modelo.score(X_train, Y_train)))
```

<h1>Aplicamos KNN o k vecinos más cercanos</h1>

```python
modelo = KNeighborsClassifier(n_neighbors=4)
modelo.fit(X_train, Y_train)
```

<h3>Predecir valores faltantes</h3>

```python
faltantes = modelo.predict(prueba)
print(faltantes)
print(quien_num.iloc[[24,25]]['Personaje'])
```

<h3>Obtener Score de predicción</h3>

```python
print("Presición Vecinos más cercanos : {}".format(modelo.score(X_train, Y_train)))
```

<h1>Aplicamos Arboles de decisión</h1>

```python
modelo = DecisionTreeClassifier()
modelo.fit(X_train, Y_train)
```

<h3>Predecir valores faltantes</h3>

```python
iter = 1000
capas_ocultas = [10,10,10]
modelo = MLPClassifier(hidden_layer_sizes = (capas_ocultas),max_iter=iter)
modelo.fit(X_train, Y_train)
```
